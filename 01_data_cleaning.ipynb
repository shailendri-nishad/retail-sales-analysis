## Step 3. Data Cleaning

### - Checking Null Values 

```python
# Checking Null Values
print(df.isnull().sum())

print('\nChecked for null values and no missing values were found.')
```

### - Fixing Column Names

```python
# Checking Column names
df.columns
```

```python
# Standardize column names for easier analysis and coding
df.columns = df.columns.str.replace(' ', '_').str.replace('-', '_').str.lower().str.strip()
```

```python
# Columns names after fixing them
df.columns
```

```python
# Dropping unnecessary Columns
df.drop('row_id', axis=1, inplace=True)
```

### - Handling duplicate records

```python
# Duplicate rows count
df.duplicated(subset=['order_id', 'product_id']).sum()
```

```python
# Show duplicate order_id + product_id records
duplicates = df[df.duplicated(subset = ['order_id', 'product_id'], keep=False)]

duplicates.sort_values(['order_id', 'product_id'])
```

```python
# Merge duplicate records and sum quantity, sales and profit values.

df = df.groupby(
    [col for col in df.columns if col not in ['quantity', 'sales', 'profit']], as_index=False).agg({
    'quantity' : 'sum',
    'sales' : 'sum',
    'profit' : 'sum'
    })
```

```python
print(f'Shape After Merging :{df.shape}')
```

### - DATA TYPE CONVERSION

```python
# Convert Date columns into Date time format

df['order_date'] = pd.to_datetime(df['order_date'], format = '%m/%d/%Y')
df['ship_date'] = pd.to_datetime(df['ship_date'], format = '%m/%d/%Y')
```

```python
# Convert categorical (object) columns into category datatype to reduce memory usage and improve analysis efficiency.

cat_cols = ['ship_mode', 'segment', 'country', 'city', 'state', 'region', 'category', 'sub_category']

for col in cat_cols:
    df[col] = df[col].astype('category')
```

```python
# Column datatypes after converting them.
df.dtypes
```

### - Creating New Features

```python
# Calculate shipping time in days
df['days_to_ship'] = (df['ship_date'] - df['order_date']).dt.days

# Extract Year
df['year'] = df['order_date'].dt.year

# Extract Month
df['month'] = df['order_date'].dt.month_name()

# Create a quarter column from Order Date
df['order_quarter'] = df['order_date'].dt.quarter
```

```python
df['profit_category'] = df['profit'].apply(lambda x: 'Profit' if x > 0 else 'Loss')
```

### - Save cleaned data in csv format

```python
df.to_csv('superstore_cleaned.csv', index=False)
print('File saved!')
```
